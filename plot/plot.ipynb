```python
import numpy as np
import matplotlib.pyplot as plt
```

## t-SNE visualization for aggregated embeddings
Using ICD code: 401.9 - Unspecified essential hypertension
- 3327 test instances
- 1441 instance with this code

```python
# import data
model_list = ["MultiResCNN-512", "MultiResCNN-2500", "BERT-512", "ClinicalBERT-512",
              "XLNet-1500", "ClinicalXLNet-1500", "Longformer-2500", "Longformer-3200"]

X_0_list = []
X_1_list = []
y_0_list = []
y_1_list = []

for model in model_list:
    X_0 = np.loadtxt(model + "_X_0.txt")
    X_1 = np.loadtxt(model + "_X_1.txt")
    X_0_list.append(X_0)
    X_1_list.append(X_1)
    y_0 = np.loadtxt(model + "_y_0.txt")
    y_1 = np.loadtxt(model + "_y_1.txt")
    y_0_list.append(y_0)
    y_1_list.append(y_1)
```

```python
fig = plt.figure(figsize=(16, 8))

for i in range(len(model_list)):
    X_0 = X_0_list[i]
    X_1 = X_1_list[i]
    y_0 = y_0_list[i]
    y_1 = y_1_list[i]
    
    ax = fig.add_subplot(2, 4, i + 1)
    ax.scatter(X_0[:, 0], X_0[:, 1], color="blue", alpha=0.5, s=2)
    ax.scatter(X_1[:, 0], X_1[:, 1], color="red", alpha=0.5, s=2)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title('(' + chr(ord('a') + i) + ") " + model_list[i])

plt.savefig("plot.pdf", bbox_inches='tight')
plt.show()
```
